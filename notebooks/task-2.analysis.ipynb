```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv(r"C:\Users\MMM\Documents\10 Academy File\KAIM-Week-7\data\telegram_data.csv")
data
```

```python
df = data
# Remove duplicates based on unique message ID
df_cleaned = df.drop_duplicates(subset='ID', keep='first')
df_cleaned
```

```python
# Check for missing values
missing_values = df_cleaned.isnull().sum()

# Handle missing values: Example, drop rows where message or date is missing
df_cleaned = df_cleaned.dropna(subset=['Message', 'Date'])
```

```python
# Convert the date column to a standardized format
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'], errors='coerce')
```

```python
import os
# Validate that each media path exists (if media is expected)
df_cleaned['Media Path Exists'] = df_cleaned['Media Path'].apply(lambda x: os.path.exists(x) if pd.notna(x) else False)
```

```python
# Store cleaned data
df_cleaned.to_csv(r'C:\Users\MMM\Documents\10 Academy File\KAIM-Week-7\data\cleaned_telegram_data.csv', index=False)
```
