```python
data["Message"]=data["Message"].fillna("No message")
i = 1
while i < len(data):
    if pd.isna(data['Media Path'].iloc[i]): 
        
        data['Media Path'].iloc[i] = data['Media Path'].iloc[i-1]
    i += 1
```

```python
data["Message"]=data["Message"].str.upper()
data["Message"]=data["Message"].str.replace(",","")
data["Message"]=data["Message"].str.replace(".","")
data["Message"]=data["Message"].str.replace("?","")
data["Message"]=data["Message"].str.replace("/","")
data["Message"]=data["Message"].str.replace("\\","")
data["Message"]=data["Message"].str.replace("-","")
import re
def remove_emojis(text):
    emoji_pattern = re.compile(
        "[" 
        "\U0001F600-\U0001F64F"  # emoticons
        "\U0001F300-\U0001F5FF"  # symbols & pictographs
        "\U0001F680-\U0001F6FF"  # transport & map symbols
        "\U0001F700-\U0001F77F"  # alchemical symbols
        "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
        "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
        "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
        "\U0001FA00-\U0001FA6F"  # Chess Symbols
        "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
        "\U00002702-\U000027B0"  # Dingbats
        "\U000024C2-\U0001F251" 
        "]+", 
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', text)

# Apply the function to the 'Message' column
data['Message'] = data['Message'].apply(remove_emojis)

# Display the updated DataFrame
data.head()
```

```python
data.dtypes
```

```python
data["Date"]=pd.to_datetime(data["Date"])
```

```python
data['valid_channel_title'] = data['Channel Title'].notna() & (data['Channel Title'].str.len() > 0)
data['valid_channel_username'] = data['Channel Username'].notna() & data['Channel Username'].str.match(r'^\w+$')
data['unique_channel_username'] = ~data['Channel Username'].duplicated()
data['valid_id'] = data['ID'].notna() & data['ID'].apply(lambda x: isinstance(x, int))
data['unique_id'] = ~data['ID'].duplicated()
data['valid_message'] = data['Message'].notna() & (data['Message'].str.len() > 0) & (data['Message'].apply(lambda x: isinstance(x, str)))
data['valid_date'] = data['Date'].notna() & (pd.to_datetime(data['Date'], errors='coerce').notna())
data['valid_date_range'] = data['Date'] <= pd.Timestamp.now(tz='UTC')
data['valid_media_path'] = data['Media Path'].notna() & data['Media Path'].str.len() > 0
data['valid_media_format'] = data['Media Path'].str.endswith(('.jpg', '.png', '.jpeg', '.gif'))
```
